```python
# Task 3: Functions on Dictionary

# Initial Dictionary
D1 = {"fb": 2000, "insta": 4000, "twitter": 1500, "youtube": 5000}
print(">>>", D1)

# 1. sum() → Sum of all values
print(">>> sum(D1.values()):", sum(D1.values()))

# 2. len() → Total number of items
print(">>> len(D1):", len(D1))

# 3. max() → Key with max value
print(">>> max(D1):", max(D1))         # by key
print(">>> max(D1.values()):", max(D1.values()))   # by values

# 4. min() → Key with min value
print(">>> min(D1):", min(D1))         # by key
print(">>> min(D1.values()):", min(D1.values()))   # by values

# 5. del → Delete specific key
del D1["twitter"]
print(">>> After del D1['twitter']:", D1)

# 6. pop() → Removes key and returns its value
removed = D1.pop("insta")
print(">>> After pop('insta'):", D1, ", Removed Value:", removed)

# 7. clear() → Clears entire dictionary
D1.clear()
print(">>> After clear():", D1)
```

```python
# Task 4: Functions to Modify Set

S1 = {"Python", "Java", "C++"}
print(">>>", S1)

# 1. add() → Adds a single element
S1.add("Ruby")
print(">>> After add('Ruby'):", S1)

# 2. update() → Adds multiple elements
S1.update(["HTML", "CSS"])
print(">>> After update(['HTML','CSS']):", S1)

# 3. clear() → Removes all elements
S2 = S1.copy()
S2.clear()
print(">>> After clear():", S2)

# 4. copy() → Copies the set
S2 = S1.copy()
print(">>> After copy():", S2)

# 5. discard() → Removes element (no error if not present)
S1.discard("Java")
print(">>> After discard('Java'):", S1)

# 6. remove() → Removes element (error if not present)
S1.remove("Python")
print(">>> After remove('Python'):", S1)
```

```python
###DAY3###
# Task 1 - Calculator with 7 operators using if, elif, else

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, %, //, **): ")

if op == '+':
    print("Result:", num1 + num2)
elif op == '-':
    print("Result:", num1 - num2)
elif op == '*':
    print("Result:", num1 * num2)
elif op == '/':
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero")
elif op == '%':
    if num2 != 0:
        print("Result:", num1 % num2)
    else:
        print("Error: Modulus by zero")
elif op == '//':
    if num2 != 0:
        print("Result:", num1 // num2)
    else:
        print("Error: Floor division by zero")
elif op == '**':
    print("Result:", num1 ** num2)
else:
    print("Invalid operator!")
```

```python
# Task 3 - Calculator using Structural Pattern Matching (match-case)

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, %, //, **): ")

match op:
    case '+':
        print("Result:", num1 + num2)
    case '-':
        print("Result:", num1 - num2)
    case '*':
        print("Result:", num1 * num2)
    case '/':
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Error: Division by zero")
    case '%':
        if num2 != 0:
            print("Result:", num1 % num2)
        else:
            print("Error: Modulus by zero")
    case '//':
        if num2 != 0:
            print("Result:", num1 // num2)
        else:
            print("Error: Floor division by zero")
    case '**':
        print("Result:", num1 ** num2)
    case _:
        print("Invalid operator!")
```

```python
# Task 4 - File Handling Concepts in Python

import os

# -------- 1. Open a file in write mode --------
f = open("demo.txt", "w")   # creates a new file if not exists
f.write("Hello World!\n")
f.write("This is a file handling example.\n")
f.write("Python makes it easy.\n")
print("File created and data written.")
f.close()

# -------- 2. Open file in read mode --------
f = open("demo.txt", "r")
print("\nReading full file using read():")
print(f.read())    # reads the whole file
f.close()

# -------- 3. Open file and read line by line --------
f = open("demo.txt", "r")
print("\nReading line by line using readline():")
print(f.readline())   # first line
print(f.readline())   # second line
f.close()

# -------- 4. Using tell() and seek() --------
f = open("demo.txt", "r")
print("\nCurrent position (tell):", f.tell())   # position of cursor
print("First 5 chars:", f.read(5))
print("Position after reading 5 chars:", f.tell())
f.seek(0)   # move cursor back to beginning
print("After seek(0), reading again:", f.readline())
f.close()

# -------- 5. Append data to file --------
f = open("demo.txt", "a")
f.write("Appending new line to the file.\n")
f.close()

# -------- 6. Check if file exists --------
if os.path.exists("demo.txt"):
    print("\nFile exists: demo.txt")

# -------- 7. Remove file --------
os.remove("demo.txt")
print("File removed successfully.")
```

```python
########DAY4########
# Task 1: Fibonacci & Factorial using recursive functions

# Recursive function for factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Recursive function for Fibonacci
def fibonacci(n):
    if n <= 0:
        return "Invalid input! Enter a positive integer."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


# Example usage in Jupyter Notebook
num = 5

print(f"Factorial of {num} is: {factorial(num)}")

print("Fibonacci sequence:")
for i in range(1, num + 1):
    print(fibonacci(i), end=" ")
```

```python
# ==============================
# TASK 3: Comprehensions & Runtime Arguments
# ==============================

# ------------------------------
# Part 1: List, Tuple & Dictionary Comprehensions
# ------------------------------

print("=== Part 1: Comprehensions ===")

# List comprehension: squares of numbers 1 to 5
list_comp = [x**2 for x in range(1, 6)]
print("List Comprehension:", list_comp)

# Tuple comprehension using generator expression (converted to tuple)
tuple_comp = tuple(x**3 for x in range(1, 6))
print("Tuple Comprehension:", tuple_comp)

# Dictionary comprehension: number and its square
dict_comp = {x: x**2 for x in range(1, 6)}
print("Dictionary Comprehension:", dict_comp)


# ------------------------------
# Part 2: Runtime Arguments with *args & **kwargs
# ------------------------------

print("\n=== Part 2: Runtime Arguments ===")

# Function with *args (variable number of positional arguments)
def add_numbers(*args):
    return sum(args)

print("*args Example:")
print("Sum of numbers:", add_numbers(10, 20, 30, 40))  # Output: 100

# Function with **kwargs (variable number of keyword arguments)
def student_info(**kwargs):
    print("\n**kwargs Example:")
    for key, value in kwargs.items():
        print(f"{key} : {value}")

student_info(Name="Harivardhini", Age=22, Course="CSE-IoT")
```

```python
# Task 2 - Inheritance Types

# 1. Multiple Inheritance
class Father:
    def skills(self):
        print("Father: Driving, Gardening")

class Mother:
    def skills(self):
        print("Mother: Cooking, Painting")

class Child(Father, Mother):
    def skills(self):
        print("Child has skills from both parents:")
        Father.skills(self)
        Mother.skills(self)

c = Child()
c.skills()


# 2. Multilevel Inheritance
class Grandparent:
    def property(self):
        print("Grandparent: Owns farmland")

class Parent(Grandparent):
    def property(self):
        print("Parent: Owns a house")

class Child2(Parent):
    def property(self):
        print("Child: Owns a car")

obj = Child2()
obj.property()


# 3. Hierarchical Inheritance
class Animal:
    def speak(self):
        print("Animals can make sounds")

class Dog(Animal):
    def speak(self):
        print("Dog says: Woof Woof")

class Cat(Animal):
    def speak(self):
        print("Cat says: Meow Meow")

d = Dog()
c = Cat()
d.speak()
c.speak()


# 4. Hybrid Inheritance
class Student:
    def study(self):
        print("Student studies subjects")

class Sports:
    def play(self):
        print("Enjoys playing sports")

# StudentAthlete inherits from both (multiple inheritance = hybrid type)
class StudentAthlete(Student, Sports):
    def routine(self):
        print("Manages both studies and sports")

sa = StudentAthlete()
sa.study()
sa.play()
sa.routine()
```

```python
# Task 3 - Method Overriding & Overloading

# Method Overriding
class Vehicle:
    def move(self):
        print("Vehicle can move")

class Car(Vehicle):
    def move(self):
        print("Car drives on roads")

class Boat(Vehicle):
    def move(self):
        print("Boat sails on water")

v = Vehicle()
c = Car()
b = Boat()

v.move()
c.move()  # overriding in action
b.move()  # overriding in action


# Method Overloading (Python does not support it directly,
# but we can achieve it using default arguments or *args)

class Calculator:
    def add(self, a=0, b=0, c=0):
        return a + b + c

calc = Calculator()
print("Add 2 numbers:", calc.add(10, 20))
print("Add 3 numbers:", calc.add(10, 20, 30))
print("Add 1 number:", calc.add(5))
```
